**Desafio 07: Deixa os gráficos bonitões. (Matplotlib.pyplot)**

```python
import matplotlib.pyplot as plt

tratamentos = dados['tratamento'].value_counts(normalize = True)

tratamentos.plot.pie(labeldistance=1.1, explode=[0, .15], textprops={"fontsize": 12})

# Garante a forma circular do grafico
plt.axis("equal")

# Cria uma legenda para cada cor
plt.legend()

plt.title('Proporção de tratamentos')
plt.ylabel('')
```
