**Desafio 02: Melhorar a visualização alterando o tamanho da fonte.**


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set() #traz uma melhoria na visualização
plt.figure(figsize = (10, 6)) #tamanho do eixo x e y no gráf.
ax = sns.countplot(x = 'composto', data = dados.query('composto in @cod_compostos'), order = cod_compostos)
ax.set_title('Top 5 compostos',fontsize = 18)
ax.set_ylabel('Contador', fontsize = 14)
ax.set_xlabel('Compostos', fontsize = 14)
plt.show()
```

**Desafio 03: Plotar os histogramas com seaborn.**

```python
plt.figure(figsize = (10, 6))
ax_hist= sns.histplot(data=dados.loc[:,'g-0':'g-771'].describe().T['mean'], bins = 100, kde=True) #kde é a densidade do plot
ax_hist.set_title('Distribuição da expressão genética G',fontsize = 18)
ax_hist.set_ylabel('Quantidade de drogas utilizadas', fontsize = 14)
ax_hist.set_xlabel('Média', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
ax_hist= sns.histplot(data=dados.loc[:,'g-0':'g-771'].describe().T['min'], bins = 30, kde=True) #kde é a densidade do plot
ax_hist.set_title('Distribuição da expressão genética G',fontsize = 18)
ax_hist.set_ylabel('Quantidade de drogas utilizadas', fontsize = 14)
ax_hist.set_xlabel('Mínimo', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
ax_hist= sns.histplot(data=dados.loc[:,'g-0':'g-771'].describe().T['max'], bins = 30, kde=True) #kde é a densidade do plot
ax_hist.set_title('Distribuição da expressão genética G',fontsize = 18)
ax_hist.set_ylabel('Quantidade de drogas utilizadas', fontsize = 14)
ax_hist.set_xlabel('Mínimo', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
ax_hist= sns.histplot(data=dados.loc[:,'c-0':'c-99'].describe().T['mean'], bins = 100, kde=True) #kde é a densidade do plot
ax_hist.set_title('Distribuição da expressão genética G',fontsize = 18)
ax_hist.set_ylabel('Quantidade de drogas utilizadas', fontsize = 14)
ax_hist.set_xlabel('Média', fontsize = 14)
plt.show()
```

**Desafio 04: Estudar sobre as estatísticas retornadas .describe()**

```python
dados.loc[:,'g-0'].describe()
```

Describe:

**count** = Contagem total das ocorrências.

**mean** = Média, soma de todos os valores dividido por sua quantidade.

**std** = Desvio padrão, é uma medida de dispersão, ou seja, é uma medida que indica o quanto o conjunto de dados é uniforme. Quando o desvio é baixo quer dizer que os dados do conjunto estão mais próximos da média.

**min** = Valor mínimo dentro das ocorrências.

**25%** = 1º quartil (Q1), é o número que deixa 25% das observações abaixo e 75% acima.

**50%** = 2º quartil (Q2) / Mediana, deixa 50% das observações abaixo e 50% das observações acima.

**75%** = 3ª quartil (Q3), deixa 75% das observações abaixo e 25% acima.

**max** = Valor máximo dentro das ocorrências.

**Desafio 05: Refletir sobre a manipulação do tamanho das visualizações.**

Uma visualização errada ou mal dimensionada pode prejudicar a análise dos dados, escondendo ou distorcendo dados.

Um gráfico plotado com uma boa dimensão ajuda na análise, permitindo visualizar pequenas diferenças de pontos que não podia ser visualizada antes com o gráfico distorcido, deixando ele bem mais objetivo para a análise que se propôs.

**Desafio 06: Fazer outras análises com o boxplot e até com o histograma.**

```python
plt.figure(figsize=(20,6))
plt.subplot(1,2,1)
ax_dose=sns.boxplot(y='g-0', x='dose', data=dados)
ax_dose.set_title('Distribuição da expressão genética G por Dose',fontsize = 16, weight='bold')
ax_dose.set_ylabel('Quantidade de drogas utilizadas', fontsize = 14, weight='bold')
ax_dose.set_xlabel('Dose', fontsize = 14, weight='bold')

plt.subplot(1,2,2)
ax_hist= sns.histplot(data=dados.loc[:,'c-0':'c-99'].describe().T['std'], bins = 100, kde=True) #kde é a densidade do plot
ax_hist.set_title('Distribuição da expressão genética G',fontsize = 18, weight='bold')
ax_hist.set_ylabel('Quantidade de drogas utilizadas', fontsize = 14, weight='bold')
ax_hist.set_xlabel('Desvio Padrão', fontsize = 14, weight='bold')

plt.show()
```
